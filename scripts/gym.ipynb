```python
import gym
import diffuser.datasets as datasets

env = gym.make("maze2d-large-v1")
```

```python
import gym
import matplotlib.pyplot as plt
import numpy as np

print(env.reset())

# rgb array to image
def plot_rgb_array(rgb_array):
    plt.figure(figsize=(5, 5))
    plt.imshow(rgb_array)
    plt.title('Maze Environment')
    plt.axis('off')  # Hide axes for a cleaner look
    plt.show()

rgb_array = env.render("rgb_array")
plot_rgb_array(rgb_array)
```

```python
env.step(action=[1.0, 0.0])
```

```python
rgb_array = env.render("rgb_array")
plot_rgb_array(rgb_array)
```

```python
result = env.step(action=[1.0, 0.0])
while result[2]!=True:
    result = env.step(action=[1.0, 0.0])
    print(result)
```
